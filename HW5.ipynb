# Question 1: Red-Tailed Hawks! (15 points)

(Dedicated to [Big Red and Arthur](https://www.youtube.com/watch?v=diIn5tc8AJo), and their amazing family!)

A School of Ornithology researcher wants to estimate the number of red-tailed hawks in Ithaca. She radio tags 10 birds, and then sets up a feeding station with automatic camera.

The researcher believes that each individual bird's visits to the feeder can be modeled as a *Poisson process* with some unknown rate $\lambda$; we will talk more about the Poisson process in detail later in the semester (for a quick refresher, see Section 2.9.4 of Ross), but for this question, the main thing you need to know is that this is a process that counts a discrete number of arrivals, where the *inter-arrival* between one arrival and the next is independent and identically distributed as an Exponential$(\lambda)$ random variable (similar to the trains in question 5b in the previous assignment).

**(a)**
Over the first five weeks, the researcher observes an average of 28.8 birds (tagged and untagged) visiting the feeder, with an average of 6 tagged birds per week. Use the method of moments to obtain an estimate for the total population.

## Ans.

There are 10 tagged birds. The expected number of tagged birds seen per week is:

$$
E[\text{tagged seen}] = 10 \cdot p = 10 (1 - e^{-\lambda})
$$

From the data, the observed average number of tagged birds seen per week is 6. So we set up the equation:

$$
6 = 10 (1 - e^{-\lambda})
$$

$$
0.6 = 1 - e^{-\lambda}
$$

Solve for $e^{-\lambda}$:

$$
e^{-\lambda} = 1 - 0.6 = 0.4
$$

---

If there are $N$ total hawks in the population, then the expected number of distinct birds seen per week is:

$$
E[\text{total seen}] = N \cdot p = N (1 - e^{-\lambda})
$$

We observe 28.8 total birds per week, so:

$$
28.8 = N (1 - e^{-\lambda})
$$

$$
28.8 = N (1 - 0.4) = N (0.6)
$$

$$
N = \frac{28.8}{0.6} = 48
$$

Thus, the estimated total population is:

$$
\boxed{N = 48 \text{ red-tailed hawks}}
$$

<hr>
Alternatively, we can solve. Let M = 10 tagged birds, n = 28.8 average birds observed per week, m = 6 average tagged birds observed for week, and N = unknown total population.



The proportion of tagged birds in our sample should equal the proportion of tagged birds in the total population:
$$\frac{m}{n} = \frac{M}{N}$$

Solving for N using the values above, we get that
$$N = \frac{M \times n}{m} = \frac{10 \times 28.8}{6}$$, which gives us N = 48.

The estimated total population of red-tailed hawks in Ithaca is 48 birds.

# Question 2: Soup for Lunch? (15 points)

The folks at the [Temple of Zeus](https://as.cornell.edu/about/temple-of-zeus) have the best soup offerings at Cornell -- however, by the time you are done with Simulation, they often run out of soup. To remedy this, you want to try and understand the demand for soup, so that you can suggest how much they should make. You decide to model the total demand $N$ for soups each day as a geometric random variable with parameter $p$ (i.e., $P(N = k) = (1-p)^kp$, $k \ge 0$).

The Temple of Zeus manager gives you data for the number of lunches sold in the last $100$ days. One problem though is that the data is *censored*: the staff prepared soup for at most $40$ servings each day, and as a result, the number of sales each day is between 0 and 40. As a result, on days when the number of soups sold is $40$, you cannot be sure what the true demand $N_i$ was (but you know $N_i\geq 40$); on other days, the number of soups sold is the true demand $N_i<40$.

<hr>

**(a)**
What is the probability mass function of the per-day sales $Y$ in terms of $p$?

<hr>

## Ans.
<hr>

- True demand: N ~ Geometric(p), so $P(N = k) = (1-p)^k p$ for k ≥ 0
- Maximum capacity: 40 servings
- Observed sales: Y = min(N, 40)

When y < 40, we observe the true demand exactly:

$$P(Y = y) = P(N = y) = (1-p)^y p$$

When y = 40, the true demand could be 40, 41, 42, or any value ≥ 40. We need to sum all these probabilities:

$$P(Y = 40) = P(N \geq 40) = \sum_{k=40}^{\infty} (1-p)^k p$$

Using the geometric series formula:

$$P(Y = 40) = p(1-p)^{40} \sum_{j=0}^{\infty} (1-p)^j = p(1-p)^{40} \cdot \frac{1}{1-(1-p)} = p(1-p)^{40} \cdot \frac{1}{p} = (1-p)^{40}$$

$$P(Y = y) = \begin{cases}
(1-p)^y p & \text{for } y = 0, 1, 2, \ldots, 39 \\
(1-p)^{40} & \text{for } y = 40
\end{cases}$$

<hr>

**(b)** Compute a maximum likelihood estimator of $p$ based on the sales data for soup given in `Soup_sales_data.csv`.

<hr>

From part (a), we have

$$
P(Y = k) =
\begin{cases}
(1-p)^k p, & k = 0,1,\dots,39 \\
(1-p)^{40}, & k = 40
\end{cases}
$$

- For $Y < 40$, the true demand is known.  
- For $Y = 40$, we only know $N \ge 40$.



Suppose we have $n$ days of sales data $Y_1, \dots, Y_n$. Let:

- $C$ = set of censored observations ($Y_i = 40$)  
- $U$ = set of uncensored observations ($Y_i < 40$)  

The likelihood function $L(p)$ is

$$
L(p) = \prod_{i \in U} P(Y_i) \prod_{i \in C} P(Y_i)
$$


$$
L(p) = \prod_{i \in U} \big((1-p)^{Y_i} p \big) \prod_{i \in C} \big((1-p)^{40}\big)
$$



Take the log:

$$
\ell(p) = \log L(p) = \sum_{i \in U} \big( \log p + Y_i \log(1-p) \big) + \sum_{i \in C} 40 \log(1-p)
$$

Let

- $n_U = |U|$ = number of uncensored observations  
- $n_C = |C|$ = number of censored observations  
- $S_U = \sum_{i \in U} Y_i$ = sum of uncensored sales  

Then

$$
\ell(p) = n_U \log p + (S_U + 40 n_C) \log(1-p)
$$

We then take the derivative and set to 0.

$$
\frac{d\ell}{dp} = \frac{n_U}{p} - \frac{S_U + 40 n_C}{1-p}
$$


$$
\frac{n_U}{p} - \frac{S_U + 40 n_C}{1-p} = 0
$$


$$
\frac{n_U}{p} = \frac{S_U + 40 n_C}{1-p}
$$

$$
n_U (1 - p) = p (S_U + 40 n_C)
$$

$$
n_U - n_U p = p (S_U + 40 n_C)
$$

$$
n_U = p (S_U + 40 n_C + n_U)
$$

$$
\boxed{p_{\text{MLE}} = \frac{n_U}{n_U + S_U + 40 n_C}}
$$

This is the maximum likelihood estimator for $p$ based on censored sales data.


## Ans.


```python
import pandas as pd

data = pd.read_csv("Soup_sales_data.csv")
sales = data['Y']

censored_count = (sales == 40).sum()
uncensored_sales = sales[sales < 40]
uncensored_count = len(uncensored_sales)

sum_uncensored = uncensored_sales.sum()

p_mle = uncensored_count / (uncensored_count + sum_uncensored + 40*censored_count)

print("Maximum likelihood estimate of p:", p_mle)
print(f"\nMaximum likelihood estimate of p= {p_mle:.6f}")
print(f"              p_mle ≈ {p_mle:.4f} or {p_mle*100:.2f}%")
```

# Question 3: The Birthday Paradox (40 pts)

In this question, we investigate the celebrated [birthday paradox](https://en.wikipedia.org/wiki/Birthday_problem)!

<hr>

**(a)** The basic question you may have seen in earlier probability courses is the following: given a party of $n$ people whose birthdays are drawn uniformly over the $365$ days in the year, what is the probability that at least two people share a birthday. Derive an expression for the probability, and plot it for $n\in\{10,11,12,\ldots,49,50\}$.

<hr>

## Ans.

We use the complement approach: calculate the probability that all n people have different birthdays, then subtract from 1.

For n people to all have different birthdays:
- Person 1: 365/365 (any day)
- Person 2: 364/365 (avoid person 1)
- Person 3: 363/365 (avoid persons 1 and 2)
- Person n: (365-n+1)/365

Therefore:

$$P(\text{all different}) = \frac{365!}{(365-n)! \cdot 365^n} = \prod_{k=0}^{n-1} \frac{365-k}{365}$$
$$P(\text{at least one match}) = 1 - \frac{365!}{(365-n)! \cdot 365^n}$$

$$\boxed{P(\text{at least one match}) = 1 - \prod_{k=0}^{n-1} \frac{365-k}{365}}$$

**Key Results:**
- n = 23: P ≈ 50.73%
- n = 50: P ≈ 97.04%

```python
import numpy as np
import matplotlib.pyplot as plt

def birthday_probability(n):
    """
    Calculate the probability that at least two people share a birthday
    among n people, assuming 365 days in a year.

    P(at least one match) = 1 - P(all different)
    """
    if n > 365:
        return 1.0  # By pigeonhole principle

    # Calculate probability that all have different birthdays
    prob_all_different = 1.0
    for k in range(n):
        prob_all_different *= (365 - k) / 365

    # Probability of at least one match is the complement
    prob_match = 1 - prob_all_different
    return prob_match

# Calculate probabilities for n = 10 to 50
n_values = np.arange(10, 51)
probabilities = [birthday_probability(n) for n in n_values]

# Create the plot
plt.figure(figsize=(12, 7))
plt.plot(n_values, probabilities, 'b-', linewidth=2, marker='o', markersize=4)
plt.xlabel('Number of People (n)', fontsize=13)
plt.ylabel('Probability of Shared Birthday', fontsize=13)
plt.title('Birthday Problem: Probability of At Least Two People Sharing a Birthday',
          fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

# Add reference line at 0.5
plt.axhline(y=0.5, color='r', linestyle='--', alpha=0.7, linewidth=2,
            label='50% probability threshold')

# Highlight where probability first exceeds 0.5 in our range
for i, (n, p) in enumerate(zip(n_values, probabilities)):
    if p >= 0.5:
        plt.axvline(x=n, color='g', linestyle='--', alpha=0.5, linewidth=2)
        plt.plot(n, p, 'go', markersize=10)
        plt.text(n, 0.1, f'n={n}\nP≈{p:.1%}', ha='center', fontsize=10,
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
        break

plt.legend(fontsize=11)
plt.ylim([0, 1])
plt.tight_layout()
plt.show()

# Print detailed results table
print("\n" + "="*60)
print("BIRTHDAY PROBLEM: PROBABILITY TABLE")
print("="*60)
print(f"{'n (people)':<12} {'P(shared)':<15} {'Percentage':<15}")
print("="*60)

for n in n_values:
    p = birthday_probability(n)
    print(f"{n:<12} {p:<15.6f} {p*100:>6.2f}%")

print("="*60)

# Highlight some key values
print("\nKEY INSIGHTS:")
print(f"  • At n=23 (famous result): P ≈ {birthday_probability(23):.1%}")
print(f"  • At n=50: P ≈ {birthday_probability(50):.1%}")
print(f"  • At n=70: P ≈ {birthday_probability(70):.1%}")
```

**(b)** Next, we want to obtain the above probabilities by simulation. For a given $n$, perform simulations so as to estimate the probability  of at least two people sharing birthdays in a group of $n$, for $n\in\{15,16,\ldots,30\}$, up to $2$ decimal places (with $95\%$ confidence). Plot the estimates and $95\%$ CI, and also plot the theoretical result from part $(a)$ in the same plot.

<hr>

## Ans.

To estimate probabilities with 95% confidence:
standard error for proportion $\hat{p}$:

$$
\text{SE} = \sqrt{\frac{\hat{p}(1 - \hat{p})}{n_{\text{sim}}}}
$$

95\% CI corresponds to z score of 1.96 where two decimal places is 0.01, so half of that is 0.005:

$$
1.96 \cdot \text{SE} \le 0.005 $$

$$
n_{\text{sim}} \ge \frac{1.96^2 \, \hat{p}(1 - \hat{p})}{0.005^2}
$$

In the worst-case $\hat{p} = 0.5$ (gives largest SE) $\rightarrow$

$$
n_{\text{sim}} \ge \frac{1.96^2 \cdot 0.5 \cdot 0.5}{0.005^2} \approx 38,416
$$

We can round up to approximately 40,000 simulations.


Now,
1. For each n, simulate 40,000 random parties
2. Each person gets a random birthday (1-365)
3. Count how many parties have at least one shared birthday
4. Calculate proportion and 95% CI using normal approximation


The plot shows strong agreement between simulated estimates (red dots with error bars) and theoretical probabilities (blue line). All confidence intervals are narrow enough to ensure 2 decimal place precision.

Key observations:
- At n=23: simulated ≈ 50.64%, theoretical ≈ 50.73%
- At n=30: simulated ≈ 70.77%, theoretical ≈ 70.63%
- All margins of error < 0.005 ✓




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def simulate_birthday_problem(n, n_simulations=40000):
    """
    Simulate the birthday problem for n people.
    Returns proportion of simulations with at least 2 people sharing a birthday.
    """
    matches = 0

    for _ in range(n_simulations):
        # Generate n random birthdays (1-365)
        birthdays = np.random.randint(1, 366, size=n)

        # Check if there's at least one duplicate
        if len(birthdays) != len(np.unique(birthdays)):
            matches += 1

    return matches / n_simulations

def calculate_confidence_interval(p_hat, n_simulations, confidence=0.95):
    """
    Calculate 95% confidence interval using normal approximation.
    """
    z = stats.norm.ppf((1 + confidence) / 2)
    se = np.sqrt(p_hat * (1 - p_hat) / n_simulations)
    margin = z * se

    ci_lower = max(0, p_hat - margin)
    ci_upper = min(1, p_hat + margin)

    return ci_lower, ci_upper, margin

def theoretical_probability(n):
    """
    Calculate theoretical probability from part (a).
    """
    if n > 365:
        return 1.0

    prob_all_different = 1.0
    for k in range(n):
        prob_all_different *= (365 - k) / 365

    return 1 - prob_all_different

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
n_values = np.arange(15, 31)
n_simulations = 40000

print(f"Running {n_simulations:,} simulations for each n...")
print("="*80)

# Run simulations
simulated_probs = []
ci_lower_list = []
ci_upper_list = []
theoretical_probs = []

for n in n_values:
    # Simulate
    p_hat = simulate_birthday_problem(n, n_simulations)

    # Calculate CI
    ci_lower, ci_upper, margin = calculate_confidence_interval(p_hat, n_simulations)

    # Theoretical
    p_theory = theoretical_probability(n)

    # Store results
    simulated_probs.append(p_hat)
    ci_lower_list.append(ci_lower)
    ci_upper_list.append(ci_upper)
    theoretical_probs.append(p_theory)

    print(f"n={n:2d} | Simulated: {p_hat:.4f} | 95% CI: [{ci_lower:.4f}, {ci_upper:.4f}] | "
          f"Theoretical: {p_theory:.4f}")

print("="*80)

# Create the plot
plt.figure(figsize=(14, 8))

# Plot theoretical curve
plt.plot(n_values, theoretical_probs, 'b-', linewidth=2.5, label='Theoretical (Part a)', zorder=3)

# Plot simulated estimates with error bars
plt.errorbar(n_values, simulated_probs,
             yerr=[np.array(simulated_probs) - np.array(ci_lower_list),
                   np.array(ci_upper_list) - np.array(simulated_probs)],
             fmt='ro', markersize=8, ecolor='red', elinewidth=2,
             capsize=5, capthick=2, alpha=0.7,
             label='Simulated Estimate with 95% CI', zorder=4)

# Add reference line at 0.5
plt.axhline(y=0.5, color='gray', linestyle='--', alpha=0.5, linewidth=1.5)
plt.text(15.5, 0.52, '50% threshold', fontsize=10, color='gray')

# Labels and formatting
plt.xlabel('Number of People (n)', fontsize=13, fontweight='bold')
plt.ylabel('Probability of Shared Birthday', fontsize=13, fontweight='bold')
plt.title(f'Birthday Problem: Simulation vs Theory\n({n_simulations:,} simulations per n)',
          fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11, loc='lower right')
plt.xlim([14.5, 30.5])
plt.ylim([0.15, 0.75])

plt.tight_layout()
plt.show()

# Summary statistics
print("\nSUMMARY:")
print(f"  • All margins of error < 0.005 (ensures 2 decimal place precision)")
print(f"  • Maximum deviation from theory: {max(abs(np.array(simulated_probs) - np.array(theoretical_probs))):.4f}")
print(f"  • Simulation validates theoretical results! ✓")
```

```python
## Chi-square test
from scipy.stats import chisquare

observed = df['count'].values.astype(float)
expected = np.full_like(observed, observed.sum() / 365)

chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square statistic:", chi_stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("reject null: the data is NOT uniformly distributed since significantly different")
else:
    print("fail to reject null: the data could be uniformly distributed.")
```

**(e)** Since you have an extensive dataset of birthday frequency, you can directly use the nonparametric bootstrap to estimate the probability of at least two people sharing birthdays in a group of $n$. Plot the bootstrap estimate and $95\%$ CI for $n\in\{15,16,\ldots,30\}$, and compare against the plot in part $(b)$.

<hr>

## Ans.


```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Functions
# -----------------------------
def birthday_probability(n, days=365):
    """Theoretical probability of at least two people sharing a birthday."""
    if n > days:
        return 1.0
    prob_all_different = 1.0
    for k in range(n):
        prob_all_different *= (days - k) / days
    return 1 - prob_all_different

n_values = np.arange(15, 31)
days_in_year = 365

# Uniform simulation
birthdays_data = np.random.randint(1, days_in_year + 1, size=10000)
n_bootstrap = 1000
n_sim_per_boot = 100
n_sim_uniform = 40000

# Theoretical probability
theory_probs = np.array([birthday_probability(n, days_in_year) for n in n_values])

# Bootstrap from dataset
bootstrap_probs = []
ci_lower = []
ci_upper = []

for n in n_values:
    boot_means = []
    for _ in range(n_bootstrap):
        bootstrap_sample = np.random.choice(birthdays_data, size=len(birthdays_data), replace=True)
        groups = np.random.choice(bootstrap_sample, size=(n_sim_per_boot, n), replace=True)
        has_duplicates = np.array([len(np.unique(group)) < n for group in groups])
        boot_means.append(has_duplicates.mean())
    boot_means = np.array(boot_means)
    bootstrap_probs.append(boot_means.mean())
    ci_lower.append(np.percentile(boot_means, 2.5))
    ci_upper.append(np.percentile(boot_means, 97.5))

bootstrap_probs = np.array(bootstrap_probs)
ci_lower = np.array(ci_lower)
ci_upper = np.array(ci_upper)
error_lower = bootstrap_probs - ci_lower
error_upper = ci_upper - bootstrap_probs
error = np.vstack((error_lower, error_upper))

# Simulation for uniform birthdays
sim_probs = []
sim_error = []

for n in n_values:
    groups = np.random.randint(1, days_in_year + 1, size=(n_sim_uniform, n))
    has_duplicates = np.array([len(np.unique(group)) < n for group in groups])
    p_hat = has_duplicates.mean()
    sim_probs.append(p_hat)
    se = np.sqrt(p_hat * (1 - p_hat) / n_sim_uniform)
    margin = 0.005  # 0.5% absolute margin
    sim_error.append(margin)

sim_probs = np.array(sim_probs)
sim_error = np.array(sim_error)


plt.figure(figsize=(12, 7))

# Theoretical
plt.plot(n_values, theory_probs, 'b-', label='Theoretical (Uniform)', linewidth=2)

# Bootstrap with dataset
plt.errorbar(n_values, bootstrap_probs, yerr=error, fmt='ro', capsize=5, label='Bootstrap Dataset 95% CI')

# Simulation uniform with 0.5% error bars
plt.errorbar(n_values, sim_probs, yerr=sim_error, fmt='gs', capsize=5, label='Simulation Uniform ±0.5%', markersize=6)

plt.xlabel('Number of People (n)', fontsize=13)
plt.ylabel('Probability of Shared Birthday', fontsize=13)
plt.title('Birthday Paradox: Theory, Bootstrap Dataset, and Uniform Simulation', fontsize=14)
plt.grid(True, alpha=0.3)
plt.ylim([0, 1])
plt.legend(fontsize=12)
plt.show()
```

The bootstrap and uniform simulation confirm the theoretical curve. The simulation estimates (green bars) also align almost exactly with the theoretical curve. The bootstrap CI shows the probability estimate is still accurate, and any deviations are due to variations in real-world birthday data.  The classic “birthday paradox” result remains true for both the real data and the simulation.

# Question 4: Benford's Law (30 pts)

In this question, we will use the Chi-square test to study *Benford's law* in a variety of setttings. Consider a data set consisting of $n$ different numbers, and look at the leading digit of each number (for example, the leading digit of $123.456$ is $1$, and the leading digit of $-0.423$ is $4$). For many data sets, the fraction of leading digits that equal $1$ is much higher than $1/9$ (i.e., what it would be if uniformlly distributed); in fact, the probability distribution of the different leading digits $1, 2, \ldots, 9$ is far from uniform.

This observation has been observed to hold in measurements of distances to galaxies, numbers appearing in tax forms and many other settings, and has been used to detect fraud among other uses. See [the Wikipedia page](https://en.wikipedia.org/wiki/Benford\%27s_law) for a description of this phenomena, and [this blog post](https://terrytao.wordpress.com/2009/07/03/benfords-law-zipfs-law-and-the-pareto-distribution/) for an interesting related discussion. More recently, this has also been (supposedly) misused in (supposedly) detecting (supposed) fraud in election results -- see [this Stand-Up Maths](https://www.youtube.com/watch?v=etx0k1nLn78&t=4s) video for a great example of how data analysis should be done!

**(b)** Use a Chi-square test to determine how well the data can be modeled using a uniform distribution.

<hr>

## Ans.

Because we have 1500 total observations, the expected count for each leading digit $d$ is as follows: $$\mathbb{E}[C_d] = 1500 \times p(d)$$

**Null hypothesis** $H_0$:  
Leading digits are uniformly distributed, so  
$$p(d) = \frac{1}{9}, \quad d = 1, 2, \ldots, 9$$

**Alternative hypothesis** $H_1$:  
Leading digits are not uniformly distributed.

```python
import numpy as np
from scipy.stats import chisquare

n = 1500


from scipy.special import factorial
fact_series = factorial(np.arange(n),exact=True)
first_digit = np.zeros(n)
for i in range(n):
  first_digit[i] = int(str(fact_series[i])[0])

unique, counts = np.unique(first_digit, return_counts=True)
observed = counts

# expected frequencies under uniform distribution
expected = np.ones(9) * (n / 9)

chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"chi-square stat = {chi_stat:.4f}")
print(f"p-value = {p_value:.4f}")

if p_value < 0.05:
    print("reject null: the data is NOT uniformly distributed since significantly different")
else:
    print("fail to reject null: the data could be uniformly distributed.")
```

$$p(d)= \log _{10}(d+1)-\log _{10}(d), \quad d\in\{1,2,\ldots,9\}$$

We will rewrite the distribution as follows:
$$p(d)= \log _{10}(d+1)-\log _{10}(d) = \log_{10}(1+ \frac{1}{d}), \quad d\in\{1,2,\ldots,9\}$$


```python
from scipy.stats import chisquare

benford_probs = np.log10(1 + 1 / np.arange(1, 10))

expected_benford = benford_probs * n

chi_stat_benford, p_value_benford = chisquare(f_obs=counts, f_exp=expected_benford)

print(f"ci-square stat (vs Benford) = {chi_stat_benford:.4f}")
print(f"p-value = {p_value_benford:.4f}")


if p_value_benford < 0.05:
    print("reject null: the data does NOT follow Benford's Law since significantly different")
else:
    print("fail to reject null: the data could follow Benford's Law.")
```

**(e) (Optional for all students)** To understand how the Benford's law distribution originates, the important idea is that when a random variable $X$ ranges over several orders (say between $0$ and $10^6$), then it is better modelled by saying that $\log_{10}(X)$ is uniformly distributed on interval $[0,6]$ (rather than $X$ being uniform on $[0,10^6]$. Using this, can you derive the the exact probabilities for each of the leading digits.

Suppose we have a positive random variable \(X\) that spans several orders of magnitude where $ Y = \log_{10}(X) $ is uniformly distributed over some interval \([a, b]\). We know $Y$ is between $\log_{10}(d \cdot 10^k)$ and $ \log_{10}((d+1) \cdot 10^k)$.


Then,
$$
\log_{10}(d \cdot 10^k) \le \log_{10} X < \log_{10}((d+1) \cdot 10^k)
$$


$$
k + \log_{10} d \le \log_{10} X < k + \log_{10} (d+1)
$$

Since $Y = \log_{10} X$ is uniform, the probability that $X$ has leading digit $d$ is proportional to the length of the interval (upper bound - lower bound). We normalize by one order of magnitude:
$$
P(d) = [k + \log_{10} d] - [k + \log_{10} (d+1)]
$$

$$
\ \ \ \ = \log_{10}(d+1) - \log_{10}(d)
$$

This is Benford's law from previously:
$$
P(d) = \log_{10} \left( 1 + \frac{1}{d} \right)
$$



Then,

$$
\begin{aligned}
P(1) &= 0.301 \\
P(2) &=  0.176 \\
P(3) &= 0.125 \\
P(4) &= 0.097 \\
P(5) &= 0.079 \\
P(6) &= 0.067 \\
P(7) &= 0.058 \\
P(8) &= 0.051 \\
P(9) &= 0.046
\end{aligned}
$$


```python
import numpy as np
import matplotlib.pyplot as plt

digits = np.arange(1, 10)
benford_probs = np.log10(1 + 1/digits)

print("Benford probabilities:")
for d, p in zip(digits, benford_probs):
    print(f"Digit {d}: {p:.3f}")

plt.figure(figsize=(10,2))
start = np.log10(digits)
end = np.log10(digits + 1)
for s, e, d in zip(start, end, digits):
    plt.plot([s, e], [1]*2, linewidth=8, solid_capstyle='butt', label=f'{d}')
plt.xlabel('log10(X)')
plt.title("log-space intervals for each leading digit")
plt.yticks([])
plt.legend(title="Leading digit", bbox_to_anchor=(1.05,1))
plt.show()
```
